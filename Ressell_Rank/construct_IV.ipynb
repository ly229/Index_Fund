```python
Russell_3 = pd.read_csv("RR_data/Russell_3_rank.csv")
```

### Concstruct the Instrument
- a binary indicator for whether a firm is assigned to the Russell 2000 Index, based on its market-cap rank near the annual June reconstitution.

```python
# Define cutoff for Russell 2000 assignment
cutoff = 1000

# Create instrument: 1 if Rank >= cutoff (assigned to Russell 2000), else 0
Russell_3['assigned_to_R2000'] = (Russell_3['Rank'] >= cutoff).astype(int)
```

#### drop rank > 5000 to avoid noise
- those firms are small and have little impact on study

```python
before = len(Russell_3)
Russell_3 = Russell_3[Russell_3['Rank'] <= 5000].copy()
after = len(Russell_3)
print(f"Dropped {before - after} rows with Rank > 5000.")
```
