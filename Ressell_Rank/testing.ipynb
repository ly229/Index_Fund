```python
import pandas as pd
```

```python
# Sample DataFrame with groups and some missing months
data = {
    'group': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C'],
    'date': ['2023-01-01', '2023-01-15', '2023-02-01', '2023-01-01', '2023-01-20', '2023-04-05', '2023-01-01', '2023-02-01'],
    'value': [10, 20, 30, 40, 50, 60, 70, 80]
}

df = pd.DataFrame(data)
df['date'] = pd.to_datetime(df['date'])

# Initialize an empty list to collect results
results = []

# Group by 'group'
df_grouped = df.groupby('group')

# Iterate over each group
for name, group in df_grouped:
    # Group by month within each group and calculate the mean
    monthly_mean = group.groupby(group['date'].dt.to_period('M'))['value'].mean().reset_index()
    
    # Convert the period back to timestamp (optional)
    monthly_mean['date'] = monthly_mean['date'].dt.to_timestamp()
    
    # Add the group name to the results
    monthly_mean['group'] = name
    
    # Append the result to the list
    results.append(monthly_mean)

# Concatenate all the results into a single DataFrame
final_result = pd.concat(results)

print(final_result)
```

```python
data = {
    'group': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C'],
    'prccd': [3, 2, 3.2, 4.1, 1.2, 2.4, 1.7, 2.8],
    'cshoc': [5.6, 6.1, 4.3, 7.2, 4.8, 6.1, 5.6, 8]
}

df = pd.DataFrame(data)

# Initialize an empty dataframe to collect results
results = pd.DataFrame()

# Group by 'group'
df_groups = df.groupby('group')

# Iterate over each group
for idx, df_group in df_groups:
    if (len(df_group)>1):
        prccd_1 = sum(df_group.prccd*df_group.cshoc)/sum(df_group.cshoc)
        prccd_2 = sum(df_group.prccd*df_group.cshoc*df_group.prccd)/sum(df_group.cshoc*df_group.prccd)
    
    # Append the result to the list
    df_group.loc[:, "prccd_1"] = prccd_1
    df_group.loc[:, "prccd_2"] = prccd_2
    
    # Concatenate all the results into a single DataFrame
    results = pd.concat([results, df_group])
    
results
```
