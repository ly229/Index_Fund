### find gvkeys with mutiple cusips in mktcap_cshoq

```python
count_df = df_mktcap_cshoq.groupby(['gvkey', 'date']).size().reset_index(name='count')

# Check for uniqueness
count_df['is_unique'] = count_df['count'] == 1

# Find pairs that are not unique
not_unique_pairs = count_df[count_df['is_unique'] == False]

# Display results
print("Pairs that are not unique:")
print(not_unique_pairs)
```

```python
gv_marked = not_unique_pairs.drop_duplicates(subset='gvkey')
print(gv_marked)
```

### find different cusips under same permco in mktcap_crsp

```python
# Group by permco and find unique ids under the same permco
df_mktcap_crsp_monthly['cusip6'] = df_mktcap_crsp_monthly['cusip'].str[:6]
df_permco = df_mktcap_crsp_monthly.groupby('permco')['cusip6'].unique().reset_index()

# Rename columns for better clarity
df_permco.columns = ['permco', 'unique_cusip']

df_multiple_cusips = df_permco[df_permco['unique_cusip'].apply(lambda x: len(x) > 1)]
```
